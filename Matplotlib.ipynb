# Histogram

```python
data=pd.read_csv('auto-mpg.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
plt.figure()
plt.hist(data['mpg'])
plt.title("Distribution of MPG")
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(6,6))
plt.hist(data['mpg'],color='red',rwidth=0.9)
plt.title('Distribution of MPG',fontsize=16)
plt.xlabel("MPG")
plt.ylabel("Frequencies")
```

```python
data.dtypes
```

```python
data[data['horsepower']=='?'].index
```

```python
data1=data.drop([32, 126, 330, 336, 354, 374])
```

```python
data1['horsepower']=data1['horsepower'].astype(float)
```

```python
data1.dtypes
```

```python
plt.hist(data1['horsepower'])
```

```python
data1[["mpg",'horsepower']].plot.hist()  # to show to graphs in same plot
```

```python
data1[['mpg','horsepower']].plot.kde()  # kde means kernel based probability distrbution
```

# scatter plot

```python
plt.scatter(data1['horsepower'],data1['mpg'],s=16,color='red',marker='^')
plt.title("Plot of HP vs MPG",fontsize=16)
plt.xlabel("horse power")
plt.ylabel('MPG')
```

```python
plt.scatter(data1['displacement'],data1['mpg'],s=12,c='black')
plt.title("Plot of displacement vs MPG",fontsize=16)
plt.xlabel('Displacement')
plt.ylabel('MPG')
```

```python
# plotting two graphs at same time

plt.figure(figsize=(8,6))
plt.subplot(2,1,1)        # 2/1 fig area, 1st position
plt.scatter(data1['horsepower'],data1['mpg'],c='red',s=12,marker='^')
plt.title("Plot of Horsepower vs MPG",fontsize=16)
plt.xlabel('Horsepower')
plt.ylabel('MPG')

plt.subplot(2,1,2)
plt.scatter(data1['displacement'],data1['mpg'],c='blue',s=12,marker='^')
plt.title("Plot of Displacement vs MPG",fontsize=16)
plt.xlabel('Displacement')
plt.ylabel('MPG')

plt.tight_layout()  # to avoid intersection of two graphs
```

# pair plots

# Heat map [for plotting correlation]

```python
sns.heatmap(slice_data.corr(),annot=True)
plt.xticks(rotation=45)         # rotates the text of x labels with 45 degree
```

# Box plot

```python
plt.boxplot(data1['mpg'])
plt.title('Box plot of MPG')
plt.xlabel('MPG')
```

```python
# Box plot using sns

sns.boxplot(x='cylinders',y='mpg',data=data1)
```

```python
sns.boxplot(x='origin',y='mpg',data=data1)
```

```python
sns.boxplot(x='origin',y='mpg',hue='cylinders',data=data1)
```
